```python
from nltk.tokenize import word_tokenize


for index, row in problems.iterrows():
    problem = str(row['Problem'])
    problem = clean.clean(problem,
      clean_all= False, # Execute all cleaning operations
      extra_spaces = True ,  # Remove extra white spaces
      stemming = False , # Stem the words
                        # If I set this True, there's a problem example: was -> wa
      stopwords = False ,# Remove stop words
                         # we need to use customized_stop_words
      lowercase = True ,# Convert to lowercase
      numbers = True ,# Remove all digits
      punct = True ,# Remove all punctuations
      # reg: str = '<regex>', # Remove parts of text based on regex
      # reg_replace: str = '<replace_value>', # String to replace the regex used in reg
      stp_lang='english'  # Language for stop words
    )

    problem = word_tokenize(problem)
    filtered_tokens = [word for word in problem if word.lower() not in customized_stop_words] # Here, I remove stop_words for each problem texts
    problem = ' '.join(filtered_tokens)
    row['Problem'] = problem # changed problem text in files with cleaned text
```

```python

#Using keywords, classify problems with DOK level here.

dok_dict = {}
processed_problems = []
def classify_problem(problem):
  for index, row in problem.iterrows():
      if any(element in row['Problem'] for element in Level1):
         dok_dict[row['Problem']] = "1"
         processed_problems.append(row['Problem'])
      elif any(element in row['Problem'] for element in Level2):
         dok_dict[row['Problem']] = "2"
         processed_problems.append(row['Problem'])
      elif any(element in row['Problem'] for element in Level3):
         dok_dict[row['Problem']] = "3"
         processed_problems.append(row['Problem'])
      elif any(element in row['Problem'] for element in Level4):
         dok_dict[row['Problem']] = "4"
         processed_problems.append(row['Problem'])
      else :
        considerProblem.append(row['Problem'])
  return dok_dict
```

```python
# This is a function for make tf-idf score for problems

def create_document_term_matrix(message_list, vectorizer):
    doc_term_matrix = vectorizer.fit_transform(message_list)
    return DataFrame(doc_term_matrix.toarray(),
                     columns=vectorizer.get_feature_names_out())
```
